```python
import cloudinary
import cloudinary.uploader
from cloudinary.utils import cloudinary_url
import os
from dotenv import load_dotenv
import json
from tqdm import tqdm
import time

load_dotenv()

CLOUD_NAME=os.environ.get("CLOUD_NAME")
API_KEY=os.environ.get("API_KEY")
API_SECRET=os.environ.get("API_SECRET")

FOLDER_PATH = "extracted_questions/images/pages"

```

```python
url_data = {}

all_files = os.listdir(FOLDER_PATH)
files = [f for f in all_files if f.endswith(".jpg")]

cloudinary.config(
    cloud_name = CLOUD_NAME,
    api_key = API_KEY,
    api_secret = API_SECRET,
    secure=True
)
```

```python
for filename in tqdm(files, desc="Uploading", unit="file"):
    filepath = os.path.join(FOLDER_PATH, filename)
    if os.path.isfile(filepath):
        url = filepath
        result = cloudinary.uploader.upload(filepath)
        url = result['secure_url']
        url_data[filename] = url
        # tqdm.write(f"uploaded: {filename}")
        # time.sleep(1)

JSON_FILE = "url_data.json"
with open(JSON_FILE, 'w') as f:
    json.dump(url_data, f, indent=4)
```

```python
import os
import json

JSON_FILE = "url_data.json"
try:
    with open(JSON_FILE, "r") as f:
        uploaded_urls = json.load(f)
    print(f"Loaded {len(uploaded_urls)} links from {JSON_FILE}")
except FileNotFoundError:
    print("Error: JSON file not found. Did you run the upload script first?")
    uploaded_urls = {}


input_md_file = "sample.md"
output_md_file = "output.md"

try:
    with open(input_md_file, 'r') as f:
        markdown_text = f.read()
except FileNotFoundError:
    print(f"ERROR: {input_md_file} not found!")

if markdown_text:
    for filename, url in uploaded_urls.items():
        markdown_text = markdown_text.replace(filename, url)
    
    with open(output_md_file, 'w') as f:
        f.write(markdown_text)
    print(f"Done! File saved as {output_md_file}.")
```
